1. (6б). Створіть клас Building, додайте три атрибути, один з яких має значення за замовчуванням та два методи на свій розсуд.

```python
class Building:
    def __init__(self, floors = 10):
        self.floors = floors
        self.number_of_floor = None
        self.rooms_per_floor = 0

    def characteristic_of_floor(self, number_of_floor, rooms_per_floor):
        self.number_of_floor = number_of_floor
        self.rooms_per_floor = rooms_per_floor

    def sum_information(self):
        quantity_buildings_rooms = self.floors * self.rooms_per_floor
        return quantity_buildings_rooms
```

3. (5б). Створіть клас, де атрибути мають різні рівні доступу. Спробуйте отримати їхні значення та опишіть результати.

```python
class A:
    def __init__(self, par1, par2, par3):
        self.param1 = par1
        self._param2 = par2
        self.__param3 = par3

    @property
    def param1(self):
        return self._param1

    @param1.setter
    def param1(self, value):
        if value < 5:
            raise ValueError("Invalid input")
        self._param1 = value

a = A(5, 2, 3)

print(a._param1)
print(a._param2)
try:
    print(a.__param3)
except AttributeError:
    print("Private variable!")
```

5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class ThreeD_Figure:
    def get_volume(self):
        return 0

    def __add__(self, other):
        return self.get_volume() + other.get_volume()
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
class Cylinder(ThreeD_Figure):
    def __init__(self, r, h, PI = 3.14159):
        self.PI = PI
        self.r = r
        self.h = h

    def get_volume(self):
        return self.PI * (self.r ** 2) * self.h

class Cube(ThreeD_Figure):
    def __init__(self, a):
        self.a = a

    def get_volume(self):
        return self.a ** 3

cylinder = Cylinder(2, 2)
cube = Cube(2)

print(f"Volume of Cube: {cube.get_volume()}")
print(f"Volume of Cylinder: {cylinder.get_volume()}")
total_volume = cube + cylinder
print(f"Total volume: {total_volume}")
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
    def __init__(self, initial_balance, wallet_id):
        self.balance = initial_balance
        self.id = wallet_id

    @property
    def balance(self):
        return self._balance

    @balance.setter
    def balance(self, money):
        if money < 0:
            raise ValueError("Balance cannot be negative!")
        self._balance = money

    @property
    def id(self):
        return self._id

    @id.setter
    def id(self, code):
        if not isinstance(code, str):
            raise ValueError("Id should be string!")
        self._id = code

    def replenishment(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Replenished with: {amount}. New balance: {self.balance}")
        else:
            raise ValueError("Amount of money cannot be negative!")

    def removal(self, amount):
        if amount > self.balance:
            raise ValueError("Insufficient funds!")
        if amount > 0:
            self.balance -= amount
            print(f"Removed: {amount}. New balance: {self.balance}")
        else:
            raise ValueError("Amount of money cannot be negative!")


wallet = EWallet(1000, "dmsldk")
wallet.replenishment(50)
wallet.removal(30)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
from typing import List

class Item:
    def __init__(self, name: str):
        self.name = name

    def __str__(self) -> str:
        return self.name

class ShoppingCart:
    def __init__(self, items: List[Item]):
        self.items = items

    def add_item(self, item: Item) -> None:
        self.items.append(item)

    def delete_item(self, item_name: str) -> None:
        self.items = [item for item in self.items if item.name != item_name]

    def __repr__(self) -> str:
        res = ", "
        item_names = [str(item) for item in self.items]
        return ", ".join(item_names)


purchase1 = Item("Birkin bag")
purchase2 = Item("Manolo shoes")
purchase3 = Item("Jane Eyre")
items = list((purchase1, purchase2, purchase3))
cart = ShoppingCart(items)

print("----- Initial basket -----")
print(cart)
cart.add_item(Item("iPhone"))
print("----- After adding -----")
print(cart)
cart.delete_item("Manolo shoes")
print("----- After deleting -----")
print(cart)
```

Ітерабельні обʼєкти - це обʼєкти, значення яких можна перебирати безліч разів і отримувати їх значення __iter__() наприклад в циклі for: tuple, str, list... Ітератори - спосіб перебирання цих обʼєктів, пам'ятаючи поточний стан, зазвичай використовується метод __next__(), використовується один раз.

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekDays:
    def __init__(self, day_of_week=0):
        self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        self.day_of_week = day_of_week

    def __iter__(self):
        return self

    def __next__(self):
        if self.day_of_week <= len(self.days):
            day = self.days[self.day_of_week - 1]
            self.day_of_week += 1
            return day
        else:
            raise StopIteration

for day in WeekDays(2):
    print(day)
```
