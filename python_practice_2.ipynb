15. Порівняйте список та кортеж. Назвіть схожості та відмінності, випадки використання.

Списки(list) та кортежі(tuple) — це схожі за своєю структурою впорядковані колекції, які дозволяють зберігати набір елементів та звертатися до них за індексом, проте між ними є відмінність: список є змінним типом даних, що дозволяє вільно додавати чи видаляти елементи, тоді як кортеж незмінний, тобто створюється один раз. Завдяки цьому кортежі працюють швидше та займають менше пам’яті, а також гарантують цілісність даних. В результаті, списки більше підходять для динамічних наборів даних, що постійно оновлюються, а кортежі — для зберігання фіксованих значень та налаштувань, які мають залишатися сталими назавжди.

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set_one = {1, 4, 6, 2, 46, 567, 34, 79, 123, 578}
set_two = {4, 57, 34, 68, 6, 3578, 24, 78, 23, 6, 46}
union_set = set_one.union(set_two)
difference_set = set_one.difference(set_two)
intersection_set = set_one.intersection(set_two)
symmetric_difference_set = set_one.symmetric_difference(set_two)
print(union_set)
print(difference_set)
print(intersection_set)
print(symmetric_difference_set)
```

# VII. Словники (Dictionary).

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
dictionary = dict()
dictionary[1] = "first"
dictionary["two"] = "second"
dictionary[('t', 'h', 'r', 'e', 'e')] = "third"
dictionary[2.454] = "fourth"

print(dictionary)
```

Ні, список не може бути ключем, через те, що ключі мають бути незмінними, а список - це змінний тип даних, тому він не може бути використаний у ролі ключа.
